```python
import pandas as pd
import os
import numpy as np

import geopandas as gpd
import xarray as xr
import rioxarray as rioxr

from shapely import Point
import matplotlib.pyplot as plt
```

```python
class_name = pd.read_csv('data/GAP_National_Terrestrial_Ecosystems.csv')
class_name
```

```python
path = os.path.join(os.getcwd(),'data','land_cover.tif')
lulc = rioxr.open_rasterio(path)

lulc
```

```python
#check type 
type(lulc)

#checking crs rios.crs

vrfe = lulc.plot.hist()
```

```python
print(lulc.dims, lulc.coords,'\n')
```

```python
#using squeeze to remove all length one, and drop removes all coordinates associated to the dimension
lulc= lulc.squeeze()
print(lulc.dims, lulc.coords,'\n')
```

```python
lulc = lulc.drop('band')
print(lulc.dims, lulc.coords)
```

```python
pixels = np.unique(lulc, return_counts = True)
```

```python
columns = {'code': pixels[0],
           'num_pixel': pixels[1]}

pixel_count = pd.DataFrame(columns)
pixel_count
```

```python
classs = pd.merge(class_name,
        pixel_count,
        on = 'code')
classs
```

```python
#store total number of pixels
total_pixel = lulc.rio.width  * lulc.rio.height
total_pixel
```

```python
percent_pixels = classs.num_pixel/total_pixel
```
